```python
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as shc
from scipy.spatial.distance import pdist

# Sample data (replace with your actual data)
data = [[1, 2], [1, 4], [3, 4], [5, 6], [7, 8], [8, 9]]

# Calculate distances between data points
distances = pdist(data)

# Perform hierarchical clustering
Z = shc.linkage(distances, method='ward')

# Create the dendrogram
plt.figure(figsize=(10, 7))
plt.title("Dendrogram")

# Plot the dendrogram
dendrogram = shc.dendrogram(Z)

# Draw horizontal lines at each node
for i, d in enumerate(sorted(dendrogram['dcoord'])[-4:]):
    plt.axhline(y = d[1], color = 'r', linestyle = '--') 


# Labels for axes
plt.xlabel('Sample index')
plt.ylabel('Distance')

# Show the plot
plt.show()
```

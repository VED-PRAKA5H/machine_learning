# Python implementation of LDA
Here's a Python code example demonstrating how to implement LDA (Linear
Discriminant Analysis) on a dataset using scikit-learn:

## Step 1: Importing necessary libraries

```python
import numpy as np
import pandas as pd
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.preprocessing import LabelEncoder, StandardScaler
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

## Step 2: Data Preprocessing and Loading

```python
iris = load_iris()
df = pd.DataFrame(data= iris.data, columns = iris.feature_names)
df['target'] = iris.target
```

### Separate Features and Target Variable:

```python
# Separate features and target variable
X = df.iloc[:,:4]
y = df.target
```

### Encode the Target Variable (if Categorical):
* We use `LabelEncoder` to convert the categorical target labels ('setosa',
'versicolor', 'virginica') into numerical format.
* This step ensures that the target variable is compatible with the LDA
algorithm.

```python
# here we don't need to do this since y is of int type
# label_encoder = LabelEncoder()
# y_encoded = label_encoder.fit_transform(y)
```

### Standardize the Features:
Standardization helps to bring all features to the same scale,\
which can
improve the performance of LDA.

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

### Instantiate LDA Object:
* We create an instance of the LDA class from scikit-learn by calling
LDA(n_components=2). Here, we specify n_components=2 to indicate that
we want to project the data onto a 2-dimensional subspace.
* Fit LDA Model: We use the fit_transform() method of the LDA object to fit the
LDA model to the standardized features (X_scaled) and the encoded
target variable (y_encoded). This method simultaneously learns the
optimal discriminant vectors and transforms the original data into the new
feature space defined by these vectors.
* Transform Features: After fitting the LDA model, we obtain the transformed
features (X_lda) by transforming the standardized features (X_scaled) into
the new feature space. Each data point in the original feature space is
projected onto the discriminant vectors, resulting in a new set of features
(LD1, LD2) that maximize class separation.

```python
lda = LDA(n_components=2)
X_lda = lda.fit_transform(X_scaled, y)
# Create a DataFrame for the transformed data
lda_df = pd.DataFrame(data=X_lda, columns=['LD1', 'LD2'])
lda_df['target'] = y.values
```

## Step 3: Plot the Transformed Data (Optional):

```python
# Plot the transformed data
plt.figure(figsize=(8, 6))
targets = df['target'].unique()
colors = ['r', 'g', 'b'] # Adjust colors if more classes
for target, color in zip(targets, colors):
    indices_to_keep = lda_df['target'] == target 
    plt.scatter(lda_df.loc[indices_to_keep, 'LD1'], lda_df.loc[indices_to_keep, 'LD2'], c=color, s=50)
plt.xlabel('LD1')
plt.ylabel('LD2')
plt.legend(['iris setosa', 'iris versicolor', 'iris virginica'])
plt.title('LDA of Iris Dataset')
plt.grid()
plt.show()
```
